```python
def clean_country_variations(s):
    """
    Returns a cleaned string based on the list of off variations, if the string name is not valid it will return an empty string
    This is a hardcoded variations list
    
    """
    variations = ('United States', 'USA', '', 'Boulder', 'United States of America',
       'Chester', 'Usa', 'United Kingdom', 'US', 'U.S.A', 'ADA', 'usa',
       '10012', 'China', 'Calhoun', 'united states', 'Cochise', 'Marin',
       'Los Angeles', 'Denver', '54481', 'South Africa', 'Canada',
       'Riverside', 'Sagadahoc', 'Moore', 'Maricopa', 'Johnson', 'USa',
       'Orange', 'Galatin', 'Santa Cruz', 'UNITED STATES', ' UNITED STATES')
    replacements = ('US', 'US', '', '', 'US', '', 'US', 'GB', 'US', 'US', '', 'US', '', 'CN', '', 'US', '', '', '', '', '', 'ZA', 'CA', '', '','','','','US','', '', '', 'US','US' )
    for i, (a,b) in enumerate(zip(variations,replacements)):
            s = s.replace((a), (b));
    return s
```
